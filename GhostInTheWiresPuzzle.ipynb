```python
from tqdm import tqdm
import bisect

def find_matching_words(file_path):
    # Open the file and read all lines
    with open(file_path, 'r') as file:
        words = file.readlines()

    hack_this = 'Wbth lal htat'
    hacking_words = hack_this.lower().split(' ')
    lengths = set([len(w) for w in hacking_words])
    by_length = {l:[] for l in lengths}
    for w in words:
        ws = w.strip()
        l = len(ws)
        if l in by_length:
            by_length[l].append(ws)
    print(by_length[3][:10])
    print('candidate list sizes')
    for l in by_length:
        print(len(by_length[l]))
    
    decrypted = []
    word2_list = by_length[len(hacking_words[2])]
    for word0 in tqdm(by_length[len(hacking_words[0])]):
        for word1 in by_length[len(hacking_words[1])]:
            word2 = word0[3]+word0[2]+word1[1]+word0[2]
            index = bisect.bisect_left(word2_list, word2)
            if index < len(word2_list):
                decrypted.append(' '.join([word0, word1, word2_list[index]]))
    print(len(decrypted))
        
    return decrypted
                    

# Example usage:
#dec = find_matching_words('/usr/share/dict/words')
dec = find_matching_words('/home/maria/Downloads/1.1million word list.txt')
```

```python
dct={' ':' ',
     'c': 'i',
     'a':'t',
     't':'a',
     'h':'d',
    }

def hypothesis(encrypted, decrypted):
    for i in range(len(encrypted)):
        dct[encrypted[i]]=decrypted[i]

hypothesis('htat','solo')
hypothesis('oy', 'of')

def decode(puzzle):
    return (''.join([dct.get(letter, letter) for letter in puzzle]),
            ''.join([dct.get(letter, '_') for letter in puzzle]))
decode(puzzle2)
```
